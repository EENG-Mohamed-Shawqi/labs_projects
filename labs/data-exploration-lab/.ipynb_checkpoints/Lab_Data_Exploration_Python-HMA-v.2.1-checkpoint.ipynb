##### 1.❓Are there better or worse days/times in terms of sales? Plot a chart to show the result

```python
# Extracting dates from transaction date
coffee_df['transaction_day']#=coffee_df['transaction_date'].dt.day because already converted
coffee_df['transaction_month']=coffee_df['transaction_date'].dt.month_name()
coffee_df['transaction_year']=coffee_df['transaction_date'].dt.year
coffee_df.head()
```

```python
# Creating time categories function to return Morning, Noon, Evenging, Night based on transaction_time
coffee_df['transaction_time']=pd.to_datetime(coffee_df['transaction_time'],format='%H:%M:%S')
def day_time(time):
    hour = time.hour
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 16:
        return 'Noon'
    elif 16 <= hour < 19:
        return 'Evening'
    else:
        return 'Night'
# use apply to apply the time category into a new column called time_category
coffee_df['time_category']=coffee_df['transaction_time'].apply(day_time)
coffee_df.head(2)
```

```python
# Are there better or worse days/times in terms of sales?
coffee_df.groupby('time_category')['line_item_amount'].sum().plot(kind='bar',figsize=(5,5),ylabel='revenue',rot=45);

# Write your findings
```

##### 3.❓Do customers who have been customers longer tend to spend more? Plot a chart to show the result (Hint : You may need to use date.today() to get todays date)

```python
# Do customers who have been customers longer tend to spend more?
from datetime import date

today = pd.to_datetime(date.today())
coffee_df['customer_loyalty_years']=round((today- coffee_df['customer_since']).dt.days/365,0)
coffee_df.groupby('customer_loyalty_years')['line_item_amount'].sum().plot(kind='bar',ylabel='revenus');
```
